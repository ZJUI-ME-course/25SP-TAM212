判断瞬心位置
---

**解题思路**  
1. 设瞬心坐标 $[x,y,z]$
2. 所有点速度为绕瞬心旋转的速度 $\vec v=\vec \omega \times \vec r$
3. 解方程

```python
omega = Matrix([0, 0, 1])
rP = Matrix([ 1, -1,  0])
vP = Matrix([ 4, -1,  0])

x,y,z=S('x,y,z')
rMP=Matrix([x,y,z])

v=omega.cross(rMP)

ans=solve([v-vP],[x,y,z])
x_v=ans[x]
y_v=ans[y]
rMP=rMP.subs(x,x_v).subs(y,y_v)
rM=rP-rMP
rM
```

已知$\vec v_Q,\vec r_{CQ}$计算角速度$\vec omega$
---

**解题思路**
1. 设瞬心M点，表示向量$\vec r_{MQ}=\vec r_{MC}+\vec r_{CQ}$
2. $\vec v_Q=\vec \omega \times \vec r_{MQ}$

**方法①使用相对速度**

```python
rCQ = Matrix([-15,   8,   0])
vQ = Matrix([25, 15,  0])

w=sy.S('w')

r=MO(rCQ)

omega=Matrix([0,0,w])
vC=Matrix([-w*r,0,0])    #注意－号
vQC=omega.cross(rCQ)

solve([vQ-vC-vQC],[w])
```

**方法②使用瞬心**

```python
rCQ = Matrix([-15,   8,   0])
vQ = Matrix([25, 15,  0])

w=sy.S('w')

r=MO(rCQ)
rMC=Matrix([0,r,0])
rMQ=rMC+rCQ
omega=Matrix([0,0,w])

vQ_com=omega.cross(rMQ)

solve([vQ_com-vQ],[w])
```

计算瞬心位置
---

```python
rP = Matrix([ 4, -5,  0])
vP = Matrix([8, 5, 0])
rQ = Matrix([3, 4, 0])
vQ = Matrix([-1,  4,  0])

z,y,z,w=sy.S('x,y,z,w')
omega=Matrix([0,0,w])
rM=Matrix([x,y,z])

rMP=rP-rM
rMQ=rQ-rM
equ1=vP-omega.cross(rMP)
equ2=vQ-omega.cross(rMQ)

solve([equ1,equ2],[x,y,z,w])
```
