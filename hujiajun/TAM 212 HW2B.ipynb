```python
#1156
#############################
t=S('t')
#############################
#此区域需要自己输入公式
x=10
y=-12
t_v=5
vr=0
ve=(-exp(3*t))
############################
r=sqrt(x**2+y**2)

#计算θ角，方便下面计算
theta=atan(abs(y/x))
if y*x<0:
    theta=pi-theta
    
if x<0 and y<0:
    theta=theta-pi

print('原坐标')
print('x',float(r*cos(theta)),'y',float(r*sin(theta)))
########################################################

#在r方向上对v积分
dr=integrate(vr,t)
r=dr+r

#对角度进行积分
de=integrate((ve/r),t)####注意角度积分的表达式
theta=theta+de
#mmd，python log 积分没有绝对值 所以遇到vr=-1,ve=-2的题目会寄

#代值
r=r.subs(t,t_v)
theta=theta.subs(t,t_v)

x=float(r*cos(theta))
y=float(r*sin(theta))

print('新坐标')
print('x',x,'y',y)
#PS 这题能不能对取决于Python计算精度
```

```python
#1057 求振幅
#本题注意事项⚠
    #y=f(x) 中变量是x不是t
    #所以对t求导的时候x=v_x*t，外面要多乘两个v_x
    
########################
x = symbols('x')
y = sin(x)
vx = 4
########################

# x = vt
#dx = vdt
ay=diff(y,x,2)*vx**2
a=ay.subs(x,pi*3.5)  #这里x带的数值要改
print(a)
```

```python
#912  解向量叉乘的方程
from numpy import *
########################
v = np.array([ 4, -1,  0])
omega = 5
########################

#设r向量
x,y,z=S('x,y,z')
r_set=np.array([x,y,z])

#将ω转化为向量形式
w=np.array([0,0,omega])

#计算v_set
v_set=np.cross(w,r_set)

#将v=v_set解方程
equ1=v[0]-v_set[0]
equ2=v[1]-v_set[1]
equ3=v[2]-v_set[2]
ans=sy.solve([equ1,equ2,equ3],[x,y,z])

x=ans[x]
y=ans[y]

print(x,y)
```

```python
#1018 (没有难度) 已知ω,r,α
##########################
x = 4
y = 0
z = 4
omega = np.array([ 3, -1,  2])
alpha = np.array([ 1, -3, -3])
##########################

r=np.array([x,y,z])
a=np.cross(alpha,r)+np.cross(omega,np.cross(omega,r))
print(a)
```

```python
#1004 (没有难度) 已知r,v求ω
##########################
x = 1
y = 5
z = -4
vMag = 6
u = np.array([ 3,  4, -4])
##########################

#单位向量化
u=u/np.linalg.norm(u)
r=np.array([x,y,z])
w=S('w')

v_set=np.cross(w*u,r)

equ=vMag**2-v_set[0]**2-v_set[1]**2-v_set[2]**2

sy.solve([equ],[w])
```
