```python
import math
from math import sin,cos,tan,asin,acos,atan,pi,sqrt
import numpy as np
import sympy as sy
x,y,z=sy.S('x,y,z')
cx,cy,bx,by=sy.S('cx,cy,bx,by')
def a2p(x):
    return x/180*pi
```

```python
#知识重点
    #1.坐标变换
    #2.投影：Orthogonal projection = 投影的向量（不是垂线）
```

```python
##T1
ax=3
ay=1

equ1=ax*cx+ay*cy-8
equ2=cx*bx+by*cy+9
equ3=ax*bx+ay*by
equ4=ax**2+ay**2-bx**2-by**2

sy.solve([equ1,equ2,equ3,equ4],[cx,cy,bx,by])
```

```python
##坐标映射
theta = -135 # in degrees
rOQ = np.array([-5, -1,  0])
rQP = np.array([ 4, -2,  0])

#向量2D化
roq=np.array([rOQ[0],rOQ[1]])
rqp=np.array([rQP[0],rQP[1]])
theta=a2p(theta)

#用i，j表示u，v
u=np.array([1*cos(-theta),1*sin(-theta)])
v=np.array([1*cos(-theta+pi/2),1*sin(-theta+pi/2)])

#用i，j表示OP向量
rop=roq+rqp

#解矩阵
A=np.array([u
           ,v])
A=A.T

np.linalg.solve(A,rop)
```

```python
theta = -50 # in degrees
r = np.array([-3, -3,  0])

theta=a2p(theta)

#用i,j表示新的基底u,v
u=np.array([1*cos(-theta),1*sin(-theta)])
v=np.array([1*cos(-theta+pi/2),1*sin(-theta+pi/2)])

#将r向量2D化
r=r[0:2]

#r=au+bv 解矩阵
A=np.array([u
           ,v])
A=A.T
np.linalg.solve(A,r)
```
