```python
#1010
import numpy as np
from sympy import *
import matplotlib.pyplot as plt
##################################################复制部分
t = symbols('t', real = True)
r = Matrix([t**2 - 3*t - 2, 2*t**2 - 2*t + 2, 0])
t_v = 3
##################################################
r=np.array([r[0],r[1]]) #三维转二维

def MO(x): #计算模长函数
    return (x[0]**2+x[1]**2)**0.5

#老白笔记计算TNB frame
T=diff(r,t,1)/MO(diff(r,t,1))
N=diff(T,t,1)/MO(diff(T,t,1))

N.subs(t,t_v)
```

```python
#该程序为 题目1010 的图像
xt=r[0]
yt=r[1]

a=-10
b=10
n=100
time=np.linspace(a,b,n)
x=np.linspace(a,b,n)
y=np.linspace(a,b,n)
for i in range(n):
    x[i]=xt.subs(t,float(time[i]))
    y[i]=yt.subs(t,float(time[i]))
plt.plot(x,y,linestyle='--')
plt.plot(xt.subs(t,t_v),yt.subs(t,t_v),'ko',linewidth='20')
```
