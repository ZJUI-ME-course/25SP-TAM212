```python
#这道题直接算就好了，就是按照他给的公式算a
r = np.array([2, 3, 0])
omega = -4


k=np.array([0,0,1])
w=k*omega

v=np.cross(w,r)
a=-sum(w*w)*r

print(v,a)
```
