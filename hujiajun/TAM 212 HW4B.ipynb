```python
from numpy import cross
#########################################复制部分
rOP = np.array([0, 1, 0])
rPQ = np.array([1, 2, 0])
omega1 = np.array([0, 0, 3])
omega2 = np.array([0, 0, 1])
alpha1 = np.array([0, 0, 3])
alpha2 = np.array([ 0,  0, -1])
#########################################计算部分

#计算P相对O点
vPO=cross(omega1,rOP)
aPO=cross(alpha1,rOP)+cross(omega1,vPO)

#计算Q相对P点
vQP=cross(omega2,rPQ)
aQP=cross(alpha2,rPQ)+cross(omega2,vQP)

aP=0+aPO
aQ=0+aPO+aQP
print('aP=',aP)
print('aQ=',aQ)
```

```python
l = 5.0
aPMag = 3.0
thetaP = 54.0 # in degrees
aQMag = 2.0
thetaQ = 76.0 # in degrees
################################复制部分
thetaP=thetaP*pi/180
thetaQ=thetaQ*pi/180
################################单位转化
a=sy.S('a')
alpha=np.array([0,0,a])
L=np.array([-l,0,0])

aQT=np.array([0,sin(thetaQ)*aQMag,0])
aPT=np.array([0,-sin(thetaP)*aPMag,0])

aPT_calculate=aQT+cross(alpha,L)

equ=aPT-aPT_calculate

print(sy.solve([equ[1]],[a]))
```

```python
from numpy import cross
#NND，这一题w1,w2,w3随机给一个，以下程序解决已知w2的情况
w1,w2,w3=sy.S('w1,w2,w3')
omega1=np.array([0,0,w1])
omega2=np.array([0,0,w2])
omega3=np.array([0,0,w3])
#########################################复制部分
rAB = np.array([2, 0, 0])
rBC = np.array([0, 1, 0])
rAD = np.array([-1,  2,  0])
rDC = np.array([ 3, -1,  0])
omega1 = np.array([ 0,  0, -3])
#########################################计算部分

#解题思路，由w1和w2算出C点速度,由w3算出C点速度，解方程
vDA=cross(omega1,rAD)
vCD=cross(omega3,rDC)
vC1=0+vDA+vCD

vC2=cross(omega2,rBC)
equ=vC1-vC2
equ1=equ[0]
equ2=equ[1]
ans=sy.solve([equ1,equ2],[w1,w2,w3])

ans
```
