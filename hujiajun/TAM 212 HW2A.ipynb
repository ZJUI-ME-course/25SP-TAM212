```python
#1097
import numpy as np
from sympy import *

#复制部分
###################################
theta = symbols('theta')

r = sin(3*theta + 3) + 3
theta_v = 2 # in radians
omega = 2
alpha = -3
###################################

#列表达式，就是求导
rx=r*cos(theta)
ry=r*sin(theta)
vx=diff(rx,theta,1)*omega
vy=diff(ry,theta,1)*omega
ax=diff(rx,theta,2)*omega**2+diff(rx,theta,1)*alpha
ay=diff(ry,theta,2)*omega**2+diff(ry,theta,1)*alpha

#带入theta_v值
r_x=float(rx.subs(theta,theta_v))
r_y=float(ry.subs(theta,theta_v))
v_x=float(vx.subs(theta,theta_v))
v_y=float(vy.subs(theta,theta_v))
a_x=float(ax.subs(theta,theta_v))
a_y=float(ay.subs(theta,theta_v))

r=np.array([r_x,r_y])
v=np.array([v_x,v_y])
a=np.array([a_x,a_y])

direction=sum(r*v)
delta_speed=sum(v*a)

if direction>0: 
    print('away')
else:
    print('toward')
    
if delta_speed>0:
    print('increase')
else:
    print('decrease')
    
    
###特别注意！！！！！！！！！！！
#这里如果先算出r向量
#再使用v=w×r的话
#算出来的只是切线速度

print('r=',r)
print('v=',v)
print('a=',a)
```
