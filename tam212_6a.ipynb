```python
from sympy import *

rCQ = Matrix([6, 8, 0])
vQ = Matrix([ 54, -18,   0])

r=rCQ.norm()
rOC=Matrix([0,r,0])

w=S('w')
omega=Matrix([0,0,w])
vc=omega.cross(rOC)
vq=omega.cross(rCQ)+vc
q1=vq-vQ
solve(q1)

```
