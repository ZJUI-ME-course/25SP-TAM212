```python
from sympy import *

rCQ = Matrix([ 4, -3,  0])
vQ = Matrix([-2,  4,  0])

#########################
k,v1=S("k,v1")

vC=Matrix([v1,0,0])
omega=Matrix([0,0,k])

q1=vQ-omega.cross(rCQ)+vC
# print(q1)
solve(q1,[k,v1])
```

```python
# 6a.4
# vp=omega.cross(rp-rm)

from sympy import *

omega = Matrix([ 0,  0, -1])
rP = Matrix([ 4, -2,  0])
vP = Matrix([-5,  3,  0])
############################
rmx,rmy=S('rmx,rmy')

rM=Matrix([rmx,rmy,0])
q1=vP-omega.cross(rP-rM)

solve(q1)
```
