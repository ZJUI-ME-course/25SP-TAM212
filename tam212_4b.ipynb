```python
# 4b.3
# vq=vp+omega2 X rPQ
# vp=omega1 X rOP
from sympy import Matrix as M
from sympy import *

rOP = Matrix([2, 3, 0])
rPQ = Matrix([1, 1, 0])
vQ = Matrix([0, 1, 0])


###############################################
w1,w2=S('w1,w2')
omega1=M([0,0,w1])
omega2=M([0,0,w2])

vq_cal=omega1.cross(rOP)+omega2.cross(rPQ)

q1=vq_cal-vQ

print(solve([q1[0],q1[1]],[w1,w2]))
```

```python
# 4b.4
# ppqx
# aq=ap+alpha.cross(rPQ)

# omega,ayalphaq1[1],ap,aq
from sympy import Matrix as M


from sympy import *

from sympy import *

from sympy import *

from sympy import *

l = 6.0
aPMag = 4.0
thetaP = 43.0 # in degrees
aQMag = 4.0
thetaQ = 57.0 # in degrees
###################################

a=S("a")
alpha=M([0,0,a])
w=S("w")
omega=M([0,0,w])

rPQ=M([l,0,0])
thetaP=thetaP/180*pi
thetaQ=thetaQ/180*pi


# pqx!!!!!!!!!!!!!!!!!!!!!!!!!!!
# pq
thetaP=thetaP
thetaQ=pi+thetaQ

ap=M([aPMag*cos(thetaP),aPMag*sin(thetaP),0])
aq=M([aQMag*cos(thetaQ),aQMag*sin(thetaQ),0])

aq_cal=alpha.cross(rPQ)+ap+omega.cross(omega.cross(rPQ))
q1=aq_cal-aq

# solution
solution = solve(q1[1], a)
print('a',solution[0])
solution2 = solve(q1[0], w)
print('w',solution2[1])
```
