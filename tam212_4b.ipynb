```python
# 1
# aop=alpha1 X rop + omega X(omega X rop)

from sympy import *

rOP = Matrix([1, 2, 0])
rPQ = Matrix([1, 0, 0])
omega1 = Matrix([0, 0, 2])
omega2 = Matrix([0, 0, 3])
alpha1 = Matrix([ 0,  0, -1])
alpha2 = Matrix([ 0,  0, -2])
#####################################################
ap=alpha1.cross(rOP)+omega1.cross(omega1.cross(rOP))

print(ap)

aq=alpha2.cross(rPQ)+omega2.cross(omega2.cross(rPQ))+ap

print(aq)

```

```python
# 2
# vq=vp+omega2 X rPQ
# vp=omega1 X rOP
from sympy import Matrix as M
from sympy import *

rOP = Matrix([2, 3, 0])
rPQ = Matrix([1, 1, 0])
vQ = Matrix([0, 1, 0])


###############################################
w1,w2=S('w1,w2')
omega1=M([0,0,w1])
omega2=M([0,0,w2])

vq_cal=omega1.cross(rOP)+omega2.cross(rPQ)

q1=vq_cal-vQ

print(solve([q1[0],q1[1]],[w1,w2]))
```

```python
# 3
# 假设ap方向是x轴


from sympy import *

l = 5.5
aPMag = 3.5
thetaP = 91.0 # in degrees
aQMag = 2.5
thetaQ = 62.0 # in degrees
thetaP=thetaP/180*pi
thetaQ=thetaQ/180*pi



ap=M([aPMag,0,0])
aq=M([aQMag*cos(thetaP-thetaQ),aQMag*-sin(thetaP-thetaQ),0])

apq=aq-ap

print(apq.evalf().norm())

```
