```python
# 4b.3
# vq=vp+omega2 X rPQ
# vp=omega1 X rOP
from sympy import Matrix as M
from sympy import *

rOP = Matrix([2, 3, 0])
rPQ = Matrix([1, 1, 0])
vQ = Matrix([0, 1, 0])


###############################################
w1,w2=S('w1,w2')
omega1=M([0,0,w1])
omega2=M([0,0,w2])

vq_cal=omega1.cross(rOP)+omega2.cross(rPQ)

q1=vq_cal-vQ

print(solve([q1[0],q1[1]],[w1,w2]))
```

```python
# 4b.4
# p点零点，然后pq方向为x负方向
# aq=ap+alpha.cross(rPQ)

# 为什么没有omega,只用a在垂直y轴方向相等，因为用的是alpha，所以最后的式子只有q1[1]是相等的,所以可以只思考ap,aq的第二项，所以我这里没有用到
from sympy import Matrix as M


from sympy import *

from sympy import *

from sympy import *

from sympy import *

l = 6.0
aPMag = 4.0
thetaP = 43.0 # in degrees
aQMag = 4.0
thetaQ = 57.0 # in degrees
###################################

a=S("a")
alpha=M([0,0,a])
w=S("w")
omega=M([0,0,w])

rPQ=M([l,0,0])
thetaP=thetaP/180*pi
thetaQ=thetaQ/180*pi


# 这里方向需要自己定，pq为x正方向!!!!!!!!!!!!!!!!!!!!!!!!!!!
# 先看p再看q
thetaP=thetaP
thetaQ=pi+thetaQ

ap=M([aPMag*cos(thetaP),aPMag*sin(thetaP),0])
aq=M([aQMag*cos(thetaQ),aQMag*sin(thetaQ),0])

aq_cal=alpha.cross(rPQ)+ap+omega.cross(omega.cross(rPQ))
q1=aq_cal-aq

# 有时候需要先定义为solution然后再输出
solution = solve(q1[1], a)
print('a是',solution[0])
solution2 = solve(q1[0], w)
print('w是',solution2[1])
```
