```python
# 4A.2
# 思路
# vpq=vq-vp
# w设成（0,0,w3）
# w X rpq = vpq 
# 只取w的w3作为body



from sympy import *

rPQ = Matrix([-4, -5,  0])
vP = Matrix([-5,  0,  0])
vQ = Matrix([ 5, -8,  0])
#########################
vq=vQ;vp=vP;rpq=rPQ;vpq=vq-vp


w3=S('w3')
w=Matrix([0,0,w3])

q1=vpq[0]-w.cross(rpq)[0]
# q2=vpq[1]-w.cross(rpq)[1]
# q3=vpq[2]-w.cross(rpq)[2]

w3=solve([q1],[w3])[w3]
print(w3)

```

```python
# 4A.3
# 思路
# w X rpq = vpq 
# vpq=vq-vp


from sympy import *

rPQ = Matrix([-4, -3,  0])
vP = Matrix([2, 4, 0])
omega = Matrix([ 0,  0, -2])

vpq=omega.cross(rPQ)
vq=vpq+vP
print(vq)
```

```python
# 4A.4
# 思路
# vpq= w X rpq
# a= alpha X rpq + w X (w X rpq)
# apq= aq-ap

from sympy import *

omega = Matrix([0, 0, 1])
alpha = Matrix([0, 0, 2])
rPQ = Matrix([-1,  1,  0])
aP = Matrix([-2,  4,  0])
###############################

vpq=omega.cross(rPQ)
apq=omega.cross(vpq)+alpha.cross(rPQ)
aq=apq+aP
print(aq)
```
