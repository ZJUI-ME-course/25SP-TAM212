```python
# 4A.4
# 思路
# vpq= w X rpq
# a= alpha X rpq + w X (w X rpq)
# apq= aq-ap

from sympy import *

omega = Matrix([0, 0, 1])
alpha = Matrix([0, 0, 2])
rPQ = Matrix([-1,  1,  0])
aP = Matrix([-2,  4,  0])
###############################

vpq=omega.cross(rPQ)
apq=omega.cross(vpq)+alpha.cross(rPQ)
aq=apq+aP
print(aq)
```

```python
# 4A.6
# 思路设w1,w2用下面两个公式表示a，然后因为是三维的，直接用公式的第一第二维即可
# va=vp+w1 X rpa
# va=vq+w2 X rqa



from sympy import Matrix as M
from sympy import *

rPA = Matrix([-1,  0,  0])
rQA = Matrix([-3, -1,  0])
vP = Matrix([-1, -2,  0])
vQ = Matrix([-3,  2,  0])
################################################

w1,w2=S('w1,w2')
omega1=M([0,0,w1])
omega2=M([0,0,w2])

q1=vP+omega1.cross(rPA)-vQ-omega2.cross(rQA)

an=solve([q1[0],q1[1]],[w1,w2])
w1=an[w1]
omega1=omega1=M([0,0,w1])
va=vP+omega1.cross(rPA)

print(va.evalf())
```
