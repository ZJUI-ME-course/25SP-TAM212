```python
# 5a.6
# 思路很简单 但是每次都需要根据不同p_positon purpose写代码手算 所以直接写一个字典来计算

omega1Mag = 0;omega2Mag = 0;omega3Mag= 0;omega4Mag = 0
alpha1Mag=0;alpha2Mag=0;alpha3Mag=0;alpha4Mag=0
w={1:0,2:0,3:0,4:0}
alpha={1:0,2:0,3:0,4:0}

from sympy import *

r1 = 8
r2 = 6
r3 = 8
r4 = 5
aPMag = 89
alpha3Mag = 6

# # 已知p所的位置！！！！！！！！！！！！！！和想求的哪个的alpha
p_position=4
purpose=2

# ################################################


key=p_position
value=purpose

if {omega1Mag,omega2Mag,omega3Mag,omega4Mag}!={0,0,0,0}:

    r={1:r1,2:r2,3:r3,4:r4}
    w={1:0,2:0,3:0,4:0}
    if omega1Mag!=0:
        w[1]=omega1Mag
        w[2]=omega1Mag*r1/r2
        w[3]=omega1Mag*r1/r3
        w[4]=omega1Mag*r1/r4
    if omega2Mag!=0:
        w[2]=omega2Mag
        w[1]=omega2Mag*r2/r1
        w[3]=omega2Mag*r2/r3
        w[4]=omega2Mag*r2/r4
    if omega3Mag!=0:
        w[3]=omega3Mag
        w[1]=w[3]*r3/r1
        w[2]=w[3]*r3/r2
        w[4]=w[3]*r3/r4
    if omega4Mag!=0:
        w[4]=omega4Mag
        w[1]=w[4]*r4/r1
        w[2]=w[4]*r4/r2
        w[3]=w[4]*r4/r3

    alpha1=S('alpha1');alpha2=S('alpha2');alpha3=S('alpha3');alpha4=S('alpha4')
    alpha={1:alpha1,2:alpha2,3:alpha3,4:alpha4}

    at=alpha[key]*r[key]
    an=w[key]**2*r[key]
    a_cal=sqrt(at**2+an**2)

    alpha[key]=solve([-a_cal+aPMag],[alpha[key]])[1][0]
    alpha[value]=alpha[key]*r[key]/r[value]

    print('已知omega，求alpha')
    print('自己判断正负')
    print(abs(alpha[value].evalf()))

elif {alpha1Mag,alpha2Mag,alpha3Mag,alpha4Mag}!={0,0,0,0}:

    r={1:r1,2:r2,3:r3,4:r4}

    alpha={1:0,2:0,3:0,4:0}

    if alpha1Mag!=0:
        alpha[1]=alpha1Mag
        alpha[2]=alpha[1]*r1/r2
        alpha[3]=alpha[1]*r1/r3
        alpha[4]=alpha[1]*r1/r4
    if alpha2Mag!=0:
        alpha[2]=alpha2Mag
        alpha[1]=alpha[2]*r2/r1
        alpha[3]=alpha[2]*r2/r3
        alpha[4]=alpha[2]*r2/r4
    if alpha3Mag!=0:
        alpha[3]=alpha3Mag
        alpha[1]=alpha[3]*r3/r1
        alpha[2]=alpha[3]*r3/r2
        alpha[4]=alpha[3]*r3/r4

    if alpha4Mag!=0:
        alpha[4]=alpha4Mag
        alpha[1]=alpha[4]*r4/r1
        alpha[2]=alpha[4]*r4/r2
        alpha[3]=alpha[4]*r4/r3

    w1,w2,w3,w4=S('w1,w2,w3,w4')
    w={1:w1,2:w2,3:w3,4:w4}

    at=alpha[key]*r[key]
    an=w[key]**2*r[key]
    a_cal=sqrt(at**2+an**2)
    w[key]=solve([a_cal-aPMag],[w[key]])[1][0]
    w[value]=w[key]*r[key]/r[value]
    print('已知omega，求alpha')
    print('自己判断正负')
    print(abs(w[value].evalf()))

```
