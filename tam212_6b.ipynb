```python
# 6b.5

from sympy import *

alpha = Matrix([0, 0, 8])
aQ = Matrix([-5, -4,  0])

########################
r=S("r")
w=S('w')
omega=Matrix([0,0,w])
rOC=Matrix([0,r,0])
rOQ=Matrix([0,2*r,0])
rCQ=Matrix([0,r,0])

ac=alpha.cross(rOC)
aq=ac+alpha.cross(rCQ)+omega.cross(omega.cross(rCQ))

print(aq)

an=solve(aq-aQ)[0]

r=an[r]
w=an[w]
print(r,w)

```

```python
from sympy import *


r = 8
aQ = Matrix([ 8, -5,  0])
##############################

w=S("w")
omega=Matrix([0,0,w])
a=S("a")
alpha=Matrix([0,0,a])

rOC=Matrix([0,r,0])
rCQ=Matrix([0,r,0])

# omega.cross(omega.cross(rOC))
ac=alpha.cross(rOC)
aq=ac+alpha.cross(rCQ)+omega.cross(omega.cross(rCQ))

print(aq)
an=solve(aq-aQ)[1]
print(an[w].evalf())
```
