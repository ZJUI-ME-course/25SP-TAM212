```python

from sympy import *

rho = 54
omega = Matrix([0, 0, 3])
alpha = Matrix([ 0,  0, -6])
rCQ = Matrix([-9, -2,  0])
R=rCQ.norm()

at=6*R
an=-9*R**2/(rho+R)
aC=Matrix([at,an,0])
aQ=aC+omega.cross(omega.cross(rCQ))+alpha.cross(rCQ)
aQ.evalf()
```

```python
from sympy import *

rho = 54
omega = Matrix([0, 0, 3])
alpha = Matrix([ 0,  0, -6])
rCQ = Matrix([-9, -2,  0])


# vc ac
# vc=-w*r*I 
# ac=-alpha*r*I-vc**2/R*J

w=omega[2]
a=alpha[2]
r=rCQ.norm()
R=r+rho
vc=-w*r
ac=Matrix([-a*r,-vc**2/R,0])
# print(ac)

aq=ac+alpha.cross(rCQ)+omega.cross(omega.cross(rCQ))
print(aq.evalf())
```

```python
from sympy import *

rho = 53
omega = Matrix([0, 0, 3])
alpha = Matrix([0, 0, 9])
rCQ = Matrix([8, 7, 0])


# vc ac
# vc=-w*r*I 
# ac=-alpha*r*I-vc**2/R*J

w=omega[2]
a=alpha[2]
r=rCQ.norm()
# 修改1
rOC=Matrix([0,-r,0])
# 修改2
R=rho-r
vc=-w*r
ac=alpha.cross(rOC)+Matrix([0,-vc**2/R,0])
aq=ac+alpha.cross(rCQ)+omega.cross(omega.cross(rCQ))
print(aq.evalf())
```
