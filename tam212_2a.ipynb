```python
# 2A.3
# 思路 
# v是对r的求导
from sympy import *
t = symbols('t', real=True)

r = Matrix([-3*cos(3*t), -exp(t), cos(3*t)])
t_v = 2
# =========================


v = r.diff(t) 
result = v.subs(t, t_v) 

vm=result.norm()
print(float(vm))
```

```python
# 2A.4
# 思路
# r是对v的积分
from sympy import *
t = symbols('t', real = True)

v = Matrix([2*t**2, -t**2 + 2*t - 2, 0])
t_v = 3
# =========================

r=integrate(v, (t,0,t_v))
print(float(r.norm()))
```

```python
# 2A.7
# 思路
# 用公式计算v,a
# ！！ away等价于r v每项相乘加和大于0 delta_speed等价于v乘a相加大约0

from sympy import *

theta=S('theta')
t=S('t')

r = 4 - 2*sin(4*theta - 3)
theta_v = 4 # in radians
omegaz = -1
alphaz = 3
# ===========================

r=Matrix([r*cos(theta),r*sin(theta),0])
v=diff(r,theta,1)*omegaz
a=diff(r,theta,2)*omegaz**2+diff(r,theta,1)*alphaz

r_=r.subs(theta,theta_v)
v_=v.subs(theta,theta_v)
a_=a.subs(theta,theta_v)

print(v_.evalf(),'\n',a_.evalf())

direction=r_.dot(v_)
delta_speed=v_.dot(a_)

if direction>0: 
    print('away')
else:
    print('toward')
if delta_speed>0:
    print('increase')
else:
    print('decrease')
    
###特别注意！！！！！！！！！！！
#这里如果先算出r向量
#再使用v=w×r的话
#算出来的只是切线速度
```
