```python
# 2A.3
# 思路 
# v是对r的求导
import numpy as np
from sympy import *

t = symbols('t', real = True)

r = Matrix([-2*cos(t), -3*t**2 - t + 1, -2*sin(2*t)])
t_v = 1
# =========================


v = r.diff(t) 
result = v.subs(t, t_v) 

vm=result.norm()
print(float(vm))
```

```python
# 2A.4
# 思路
# r是对v的积分
import numpy as np
from sympy import *

t = symbols('t', real = True)

v = Matrix([-t**2 + 2*t, -3*t**2 + 2*t - 1, 0])
t_v = 1
# =========================


r=integrate(v, (t,0,t_v))
print(float(r.norm()))
```

```python

# 2a.7

# ！！ away等价于r v每项相乘加和大于0 delta_speed等价于v乘a相加大约0

import numpy as np
from sympy import *

theta = symbols('theta')

r = 4 - 2*sin(5*theta - 4)
theta_v = -1 # in radians
omegaz = 2
alphaz = 2

R=Matrix([r*cos(theta),r*sin(theta),0])
v=R.diff(theta,1)*omegaz
a=R.diff(theta,2)*omegaz**2+R.diff(theta,1)*alphaz

r1=R.subs(theta,theta_v)
v1=v.subs(theta,theta_v)
a1=a.subs(theta,theta_v)

print(v1.evalf())

print(a1.evalf())


if r1.dot(v1)>0:
    print('away')
else:
    print('towards')

if v1.dot(a1) >0:
    print('increasing')
else:
    print('decreasing')
```

```python
# 2A.7
# 思路
# 用公式计算v,a

from sympy import *

theta=S('theta')
t=S('t')

r = 4 - 2*sin(4*theta - 3)
theta_v = 4 # in radians
omegaz = -1
alphaz = 3
# ===========================

r=Matrix([r*cos(theta),r*sin(theta),0])
v=diff(r,theta,1)*omegaz
a=diff(r,theta,2)*omegaz**2+diff(r,theta,1)*alphaz

r_=r.subs(theta,theta_v)
v_=v.subs(theta,theta_v)
a_=a.subs(theta,theta_v)

print(v_.evalf(),'\n',a_.evalf())

direction=r_.dot(v_)
delta_speed=v_.dot(a_)

if direction>0: 
    print('away')
else:
    print('toward')
if delta_speed>0:
    print('increase')
else:
    print('decrease')
    
###特别注意！！！！！！！！！！！
#这里如果先算出r向量
#再使用v=w×r的话
#算出来的只是切线速度


```
