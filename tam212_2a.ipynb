```python
# 7
# 思路
# 公式算v,a
# ！！ away等价于r v每项相乘加和大于0 delta_speed等价于v乘a相加大约0

import numpy as np
from sympy import *

theta=S('theta')
t=S('t')

r = 4 - 2*sin(4*theta - 3)
theta_v = 4 # in radians
omegaz = -1
alphaz = 3
# ===========================


r=Matrix([r*cos(theta),r*sin(theta),0])
v=diff(r,theta,1)*omegaz
a=diff(r,theta,2)*omegaz**2+diff(r,theta,1)*alphaz

r_=r.subs(theta,theta_v)
r_x,r_y,r_z=[float(r_[0]),float(r_[1]),float(r_[2])]
v_=v.subs(theta,theta_v)
v_x, v_y, v_z = [float(v_[0]), float(v_[1]), float(v_[2])]
a_=a.subs(theta,theta_v)
a_x,a_y,a_z=[float(a_[0]),float(a_[1]),float(a_[2])]


r=np.array([r_x,r_y])
v=np.array([v_x,v_y])
a=np.array([a_x,a_y])

print(v,'\n',a)
direction=sum(r*v)
delta_speed=sum(v*a)

if direction>0: 
    print('away')
else:
    print('toward')


if delta_speed>0:
    print('increase')
else:
    print('decrease')
    
###特别注意！！！！！！！！！！！
#这里如果先算出r向量
#再使用v=w×r的话
#算出来的只是切线速度


```
