```python
# 8
import numpy as np
from sympy import symbols, sin, cos, diff

theta = symbols('theta')


# ============================================
r = 4 - sin(3*theta + 2)

# ============================================


dr_dtheta = diff(r, theta)
# print(dr_dtheta)


# ====================================
theta_v = 4 # in radians
omegaz = -3
# ====================================

r_value = float(r.subs(theta, theta_v))
print(r_value)

dr_dtheta_value = float(dr_dtheta.subs(theta, theta_v))

print(dr_dtheta_value)


cos_theta = np.cos(theta_v)
sin_theta = np.sin(theta_v)

vx = dr_dtheta_value * cos_theta - r_value * omegaz * sin_theta
vy = dr_dtheta_value * sin_theta + r_value * omegaz * cos_theta

print(vx,'\n')  
print(vy)  # 输出：v_y = 4.78 m/s
```

```python
import numpy as np
from math import *
theta_val = 4  # 单位：弧度
omega_z = -3    # 单位：rad/s

# 计算 dr/dθ 和 r
dr_dtheta = -3 * np.cos(3 * theta_val + 2)
r = 4 - np.sin(3 * theta_val + 2)

# 速度分量
vr = dr_dtheta * omega_z
vtheta = r * omega_z

# 速度大小
v_mag = np.sqrt(vr**2 + vtheta**2)

print(f"径向速度 v_r = {vr:.4f} m/s")
print(f"横向速度 v_θ = {vtheta:.4f} m/s")
print(f"速度大小 |v| = {v_mag:.2f} m/s")


print(sqrt(7.6369442814**2+4.9698633024**2))
```
